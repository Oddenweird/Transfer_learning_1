```python
from transformers import pipeline

# Загрузка конвейера классификации с указанной моделью
classifier = pipeline("text-classification", model="tabularisai/multilingual-sentiment-analysis")

# Тексты для классификации
texts = [
    "Отличный товар",
    "Мне не понравилось",
    "Прекрасно выглядит, ничего не меняйте!",
    "Ничего не понял",
    "У меня не работала розетка, в остальном все отлично"
]

# Классификация текстов
results = classifier(texts)

# Печать результатов
for text, result in zip(texts, results):
    print(f"Text: {text}\nSentiment: {result['label']}, Score: {result['score']:.4f}\n")
```
